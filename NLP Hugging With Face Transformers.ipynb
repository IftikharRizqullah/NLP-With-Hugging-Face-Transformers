```python
# Import the pipeline library from transformers
from transformers import pipeline

# Initialize the fill-mask pipeline with a specific model
unmasker = pipeline("fill-mask", model="bert-base-uncased")

# Refine the input text with [MASK] for the BERT model
input_text = "In this tutorial, you will gain insights into [MASK] models and applications."
result = unmasker(input_text, top_k=2)  # Retrieve the top 2 predictions for the masked word

# Display the input text and fill-mask results
print("Input Text:", input_text)
for i, prediction in enumerate(result, 1):
    print(f"Prediction {i}: {prediction['sequence']} (Score: {prediction['score']:.4f})")
```

```python
# Import the pipeline library from transformers
from transformers import pipeline

# Initialize the NER pipeline with grouped entities enabled
ner = pipeline("ner", grouped_entities=True)

# Input text for named entity recognition
text = "My name is Iftikhar and I work at Hugging Face in Brooklyn."
result = ner(text)

# Display the NER results
print("Input Text:", text)
for entity in result:
    print(f"Entity: {entity['entity_group']}, Text: {entity['word']}, Score: {entity['score']:.4f}")
```

```python
# Import the pipeline library from transformers
from transformers import pipeline

# Initialize the question-answering pipeline with a specified model
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# Modify the context and question in Indonesian
context = "Matahari adalah bintang di pusat Tata Surya. Matahari menyediakan cahaya dan panas yang penting untuk kehidupan di Bumi."
question = "Apa itu Matahari?"

# Perform question-answering based on the modified context and question
result = qa_pipeline(question=question, context=context)

# Display the answer
print(f"Jawaban: {result['answer']}")
```

```python
# Import the pipeline library from transformers
from transformers import pipeline

# Define a function for question-answering with context and question as parameters
def answer_question(context, question):
    qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
    result = qa_pipeline(question=question, context=context)
    return result

# Define the context and question
context = (
    "Raden Adjeng Kartini (lahir 21 April 1879) adalah seorang pahlawan nasional Indonesia yang dikenal "
    "sebagai pelopor emansipasi wanita di Indonesia. Kartini berjuang untuk kesetaraan pendidikan bagi "
    "perempuan, dan ide-idenya dituangkan dalam surat-surat yang ia tulis kepada teman-temannya di Belanda."
)
question = "Siapa Kartini?"

# Get the answer using the function
result = answer_question(context, question)

# Display the answer with score
print(f"Pertanyaan: {question}")
print(f"Jawaban: {result['answer']} (Kepercayaan: {result['score']:.2f})")
```

```python
# Import the pipeline library from transformers
from transformers import pipeline

# Define a function for sentiment analysis
def analyze_sentiment(text):
    classifier = pipeline("sentiment-analysis")
    result = classifier(text)
    return result

# Define the input text for sentiment analysis
input_text = "I was really impressed by the model's performance! I would definitely recommend it to others."

# Get the sentiment analysis result using the function
result = analyze_sentiment(input_text)

# Display the sentiment result
print("Input Text:", input_text)
for sentiment in result:
    print(f"Sentiment: {sentiment['label']}, Confidence: {sentiment['score']:.2f}")
```

```python
# Import the pipeline library from transformers
from transformers import pipeline

# Define a function for summarizing text
def summarize_text(text, max_length=100, min_length=50):
    summarizer = pipeline("summarization")
    summary = summarizer(text, max_length=max_length, min_length=min_length, truncation=True)
    return summary[0]['summary_text']

# Define the input text to summarize
input_text = """
    Climate change is one of the most pressing issues facing our planet today. It refers to long-term changes
    in temperatures and weather patterns, primarily caused by human activities such as burning fossil fuels,
    deforestation, and industrial processes. These activities release large amounts of greenhouse gases into
    the atmosphere, trapping heat and leading to a gradual increase in global temperatures.

    The effects of climate change are widespread and severe, including rising sea levels, more frequent and
    intense natural disasters, disruptions to ecosystems, and adverse impacts on agriculture and food security.
    Addressing climate change requires global cooperation and significant changes to energy production,
    consumption, and land management practices. Transitioning to renewable energy sources, improving energy
    efficiency, and protecting forests are some key actions to mitigate its impact.
"""

# Get the summary using the function
summary = summarize_text(input_text, max_length=80, min_length=40)

# Display the summary
print("Summary:", summary)
```
